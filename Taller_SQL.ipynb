Listar las temporadas en que un equipo visitante anotó 7 o más goles

```python
query = """
        SELECT season, home_team_goal, away_team_goal,
            t_home.team_long_name AS home_team,
            t_away.team_long_name AS away_team
        FROM
            Match
        JOIN
            Team t_home ON Match.home_team_api_id = t_home.team_api_id
        JOIN
            Team t_away ON Match.away_team_api_id  = t_away.team_api_id
        WHERE
            away_team_goal >= 7;
"""

season_awaygoal_7 = pd.read_sql(query, conn)
season_awaygoal_7
```

Traer el Id de los equipos que desarrollan un dribbling normal y su velocidad de juego es balanceada

```python
query = """
        SELECT
          DISTINCT(Team.team_long_name) AS team,
          buildUpPlayDribblingClass, buildUpPlaySpeedClass, date
        FROM
          Team_Attributes
        JOIN
          Team ON Team_Attributes.team_api_id = Team.team_api_id
        WHERE
          buildUpPlayDribblingClass = 'Normal' AND
          buildUpPlaySpeedClass = 'Balanced';
"""
team_att_and = pd.read_sql(query, conn)
team_att_and
```

Traer el Id de los equipos que desarrollan un juego de pase largo o su velocidad de juego es rápida

```python
query = """
        SELECT
          DISTINCT(Team.team_long_name) AS team,
          buildUpPlayPassingClass, buildUpPlaySpeedClass, date
        FROM
          Team_Attributes
        JOIN
          Team ON Team_Attributes.team_api_id = Team.team_api_id
        WHERE
          buildUpPlayPassingClass = 'Long' AND
          buildUpPlaySpeedClass = 'Fast';
"""
team_att_and = pd.read_sql(query, conn)
team_att_and
```

Listar el id de los 10 jugadores con mayor estatura

```python
query = """
        SELECT
            id, player_name,
            height AS height_cm
        FROM
            Player
        ORDER BY height DESC
        LIMIT 10;
"""

players_height_10most = pd.read_sql(query, conn)
players_height_10most
```

Listar el id de los 10 jugadores con mayor estatura y ordenarlos alfabéticamente pero de forma descendente
